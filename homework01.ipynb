```
import pyspark
from pyspark.sql import SparkSession
from pyspark import SparkConf, SparkContext
import collections
```

```

conf = SparkConf().setMaster('local').setAppName('wordCounting')
sc = SparkContext.getOrCreate(conf=conf)
words = sc.textFile("input.txt")
#####
test = words.collect()
for words in test:
	print(words)
```

```
words = sc.textFile("input.txt").flatMap(lambda line: line.split(" "))
print(words.collect())
#
def lower_clean_str(x):
  punc='!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~-'
  lowercased_str = x.lower()
  for ch in punc:
    lowercased_str = lowercased_str.replace(ch, '')
  return lowercased_str
words = words.map(lower_clean_str)
print(words.collect())
```

```
total_word = words.map(lambda word: 1).reduce(lambda a, b: a + b)
print(total_word)
```

```
word_counts = words.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)

print(word_counts.collect())
```

```
output = word_counts.sortBy(lambda x: x[1], ascending=False).take(20)
for (word, count) in output:
  print(word, count)
```
